```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
```

### Plotting box and wiskle chart

```python
df = pd.DataFrame(np.random.randint(1,1000,1000),columns=['values'])
```

```python
df.plot.box(figsize=(10,5))
plt.ylabel('random values')
plt.show()
```

### Frequency tables

```python
pd.cut(df['values'],10).value_counts()
```

### Histogram

```python
df.plot.hist(figsize=(10,5),bins=20)
plt.ylabel('random values')
plt.show()
```

### Density

```python
df.plot.density(figsize=(10,5))
plt.ylabel('random values')
plt.show()
```

### Bar chart

```python
plt.figure(figsize=(10,5))
plt.bar(['Carrier','ATC','Security','Inbound'],[23,30,1,40])
plt.show()
```

```python
import seaborn as sns
```

### Heatmap

```python
corr = pd.read_csv('correl.csv',index_col=['INDEX'])
```

```python
sns.heatmap(corr,vmin=0,vmax=1,cmap=sns.diverging_palette(20,200,as_cmap=True))
plt.show()
```

### Scatter plot

```python
corr.plot.scatter(x='T',y='CTL')
plt.show()
```

### Hexagonal chart

```python
x = np.random.randint(1,1000,1000)
df = pd.DataFrame(
    {"x":x,"y":x*np.random.random(1000)}
)
df.plot.hexbin(x='x',y='y',figsize=[10,5],gridsize=20,ylabel='y variable')
plt.show()
```

### Kdplot
_this is similar to a topographical map of the relationship between two variables_

```python
plt.figure(figsize=(10,5))
sns.kdeplot(x=df.x,y=df.y,shade=True)
plt.show()
```

### Violin plots

```python
df = pd.DataFrame(
    {'x':['company %d' %(y) for x in range(100) for y in range(10)],
     'y':np.random.randint(1,500,1000)})
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x=df.x,y=df.y,inner='quartile',color='white')
plt.xticks(rotation=90)
plt.show()
```

### Visualize boxplots in different subplots

```python
comp = 0
fig, ax = plt.subplots(2,5,figsize=(15,6))
for r in range(2):
    for c in range(5):
        ax[r,c].boxplot(df[df['x'] == 'company %d' %(comp)]['y'])
        ax[r,c].set_title('company %d' %(comp))
        comp += 1
plt.show()
```
