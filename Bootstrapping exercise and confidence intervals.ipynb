```python
import pandas as np
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vals = np.random.randint(0,1000,1000)
```

```python
plt.hist(vals,density=True,bins=20)
plt.show()
```

```python

sample_means = np.random.choice(vals,size=(1,10000,5),replace=True)[0].mean(1)
plt.figure(figsize=(10,5))
sns.histplot(sample_means)
# for i in np.percentile(sample_means,[2.5,70]):
#     plt.annotate(f"cint {i:0.2f}",xy=(i,i))
#     print(i,i)
plt.fill_betweenx([0,sample_means.max()],np.percentile(sample_means,2.5),np.percentile(sample_means,97.5),color='g',alpha=0.4)
plt.show()
```
